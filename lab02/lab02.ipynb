# Laboratório 2
## Tarefa 1

### Carregue o nó de Tenda dos Milagres (da versão em inglês Tent of Miracles) em Python e mostre o conteúdo JSON.

```python
#Request info for Tent of Miracles (novel) as jsod from DBPedia
url_template_dbpedia = "http://dbpedia.org/data/{concept}.{format}"

concept = "Tent of Miracles (novel)" #Nome do livro com espaços
format = "jsod"

concept_formated = concept.replace(" ", "_")
url_dbpedia = url_template_dbpedia.replace("{concept}", concept_formated)\
                                  .replace("{format}", format)

data_dbpedia = requests.get(url_dbpedia)
js_dbpedia = data_dbpedia.json()
print(json.dumps(js_dbpedia, indent=3))
```

## Tarefa 2
### Recupere via Python e mostre o autor de Tenda dos Milagres?



```python
#Get author node from novel json
author = js_dbpedia["d"]["results"][0]["http://dbpedia.org/ontology/author"]["__deferred"]["uri"]
print("Autor de " + concept + ": " + author)
```

## Tarefa 3
### Quais os termos usados para Aspirina (consulta JSON/MeSH em Python)?


```python
#Request info for condition/medicine as json from MeSH and print it, then isolate it's code
url_template_mesh = "https://id.nlm.nih.gov/mesh/lookup/descriptor?label={concept}&match={match}&limit={limit}"

concept_mesh = "Aspirin"
match = "exact"
limit = "10"

concept_mesh_formated = concept_mesh.replace(" ", "%20")

url_mesh = url_template_mesh.replace("{concept}", concept_mesh_formated)\
                            .replace("{match}", match)\
                            .replace("{limit}", limit)

#GET data from db
data_mesh = requests.get(url_mesh)
json_mesh = data_mesh.json()

#Get concept code from uri
uri_mesh = json_mesh[0]["resource"]
last_bar = str.rindex(uri_mesh, "/")
mesh_code = uri_mesh[last_bar+1:]

#print info
print("JSON info for " + concept_mesh + ": ")
print(json.dumps(json_mesh, indent=3))
print("\nCode of " + concept_mesh + " : " +mesh_code)

```

### Escolha algum lugar no OpenStreetMap e elabore a mesma sequência de três chamadas de serviços para o seu conceito.

```python
#Request some location pointer from OpenStreetMap as json and print it
url_template_osm = "http://nominatim.openstreetmap.org/search?format={format}&q={query}"

query_osm = "Vanuatu"
format_osm = "json"

url_osm = url_template_osm.replace("{query}", query_osm)\
                          .replace("{format}", format_osm)

data_osm = requests.get(url_osm)
json_osm = data_osm.json()
print(json.dumps(json_osm, indent=3))
```

```python
#Get location id from json and print it
id_osm = json_osm[0]["osm_id"]
print(id_osm)
```

```python
#GET location info as json from OpenStreetMap using location's id
url_template_osm_info = "https://api.openstreetmap.org/api/0.6/{element_type}/{id}.{format}"

element_type = "relation"
id_osm = str(json_osm[0]["osm_id"])
format = "json"

url_osm_info = url_template_osm_info.replace("{element_type}", element_type)\
                  .replace("{id}", id_osm)\
                  .replace("{format}", format)

data_osm_info = requests.get(url_osm_info)
location_info = data_osm_info.json()
print(json.dumps(location_info, indent=3))
```

```python
#Extract wikipedia link
wikipedia_page = location_info["elements"][0]["tags"]["wikipedia"]
print(wikipedia_page)
```
